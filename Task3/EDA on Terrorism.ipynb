# Visualizing Data

## 1.Year Wise Attacks

```python
attacks=terror["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
```

```python
attacks.plot(kind="bar",color="cornflowerblue",figsize=(15,6),fontsize=13)
plt.title("Timeline of Attacks",fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.show()
```

### a.Casualities each year

```python
yc = terror[["Year","Casualities"]].groupby("Year").sum()
yc.head()
```

```python
yc.plot(kind="bar",color="cornflowerblue",figsize=(15,10))
plt.title("Year wise Casualties",fontsize=13)
plt.xticks(fontsize=14)
plt.xlabel("Years",fontsize=13)
plt.ylabel("Number of casualties",fontsize=13)
plt.show()
```

### b.Killed Each Year

```python
yk = terror[["Year","Killed"]].groupby("Year").sum()
yk.head()
```

```python
yk.plot(kind="bar",color="cornflowerblue",fontsize=13,figsize=(15,10))
plt.xlabel("Years",fontsize=15)
plt.ylabel("Number of Deaths",fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```

### c.Wounded each year

```python
yw = terror[["Year","Wounded"]].groupby("Year").sum()
yw.head()
```

```python
yw.plot(kind="bar",color="cornflowerblue",figsize=(15,6),fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.ylabel("Number of Wounded",fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```

## 2.Region wise attack

### a. Distribution of terrorist attacks over Regions from 1970-2017

```python
reg = pd.crosstab(terror.Year,terror.Region)
reg.head()
```

```python
reg.plot(kind="area",stacked=False, alpha=1.0,figsize=(20,10))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

### b.Total terrorist attacks in each region from 1970-201

```python
regt = reg.transpose()
regt["Total"]=regt.sum(axis=1)
ra = regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(15,6))
plt.title("Total Number of Attacks in each region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

### c.Total Casualties (Killed + Wounded) in eaqch region

```python
rc = terror[["Region","Casualities"]].groupby("Region").sum()
rc.head()
```

```python
rc.sort_values(by="Casualities", ascending=True)
rc
```

```python
rw = terror[["Region","Wounded"]].groupby("Region").sum().sort_values(by="Wounded",ascending=False)
rw
```

```python
rw.plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.xlabel("Regions",fontsize=15)
plt.ylabel("Number of people Killed",fontsize=15)
plt.title("People killed in each region")
plt.show()
```

```python
rc.plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.xlabel("Regions",fontsize=15)
plt.ylabel("Number of Casualties",fontsize=15)
plt.title("Casualties in each region")
plt.show()
```

# 3.Country wise attack

# a.Number of attacks in each country

```python
ct = terror["Country"].value_counts().head(10)
ct
```

```python
ct.plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.title("Country wise attacks",fontsize=13)
plt.xlabel("Countries",fontsize=13)
plt.ylabel("Number of Attacks",fontsize=13)
plt.show()
```

### b.Killed in each Country

```python
cnk = terror[["Country","Killed"]].groupby("Country").sum().sort_values(by="Killed",ascending=False)
cnk.head()
```

```python
cnk[:10].plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.title("People killed in each country",fontsize=13)
plt.xlabel("Countries",fontsize=13)
plt.ylabel("Number of Deaths",fontsize=13)
plt.show()
```

```python
cnw = terror[["Country","Wounded"]].groupby("Country").sum().sort_values(by="Country",ascending=False)
cnw.head()
```

```python
cnw[:10].plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.title("People wounded in different countries",fontsize=13)
plt.xlabel("Countries",fontsize=13)
plt.ylabel("Number of People wounded",fontsize=13)
plt.show()
```

# 4.City wise attacks

### a.Number of attacks in each city

```python
city = terror["City"].value_counts()[1:11]
city
```

### b.Total casualties (Killed+Wounded) in each city

```python
cc=terror[["City","Casualities"]].groupby("City").sum().sort_values(by="City",ascending=True).drop("Unknown")
cc.head(10)
```

```python
cc[:10].plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.title("Casualties in different Cities",fontsize=13)
plt.xlabel("Cities",fontsize=13)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()
```
